# Breast Cancer Dataset

```python
import numpy as np 
import pandas as pd 
```

Read Data

```python
breast_cancer = './input/dat'
data = pd.read_csv(breast_cancer, sep = " ", header=None)
```

Show top 10 entries

```python
data.head(10)
```

How many samples?

```python
data.shape
```

Drop first column

```python
data =data.drop(0, axis = 1)
data.head(20)
```

Extract Labels: -1 healthy, +1 malignant

```python
y = data[data.columns[0]]-3
y.head(20)
```

Extract Features

```python
X =data.drop(1, axis = 1)
X.head(20)
```

How many malignant samples?

```python
y.loc[y == 1]
```

Split Data into Test and Training Set

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression




X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = True)

print(np.shape(X_train))
print(np.shape(y_test))


```

Choose Model (logistic regression)

```python
clf = LogisticRegression()
```

Fit model to data

```python

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)


```

Display results

```python
print("Actual breast cancer : ")
print(y_test.values)

print("\nPredicted breast cancer : ")
print(y_pred)

print("\nAccuracy score : %f" %(accuracy_score(y_test, y_pred) * 100)) #fraction of correct predictions
print("Recall score : %f" %(recall_score(y_test, y_pred) * 100)) #true positives/(true positives + false negatives)
print("ROC score : %f\n" %(roc_auc_score(y_test, y_pred) * 100)) # 'receiver operating characteristic'
print(confusion_matrix(y_test, y_pred)) #C_{i,j} = number of samples in group i, predicted in group j  
```
